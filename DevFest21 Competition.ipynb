### Observing some correlations

```python
ndf=df.copy()
```

```python
ndf=ndf.drop(["id", "OperatingSystems", "Region"], axis=1 )
ndf.head()
```

```python
#defining plot_correlation and giving our dataframe as parameter
def plot_correlation_map(ndf):
#defining corr as the corelation between the df's elements
    corr = ndf.corr()
# this is to make subplots with the size 12 10 
    s , ax = plt.subplots( figsize =( 12 , 10 ) )
# this is to setup the figur's colors
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
# here we created the heatmap s
    s = sns.heatmap(
# it takes as parameters:
# a callback for the corelation function
        corr, 
# calling the colors palette we defined
        cmap = cmap,
#setting the shape of the elements to squares
        square=True, 
#setup the hight of the colors of the progress bar
        cbar_kws={ 'shrink' : .9 }, 
# setting x/y axix and giving them the types of the data frame as values
        ax=ax, 
# setting annot to true to display the values on the heatmap
        annot = True, 
# this is to setup the font size
        annot_kws = { 'fontsize' : 10 }

        )
```

```python
plot_correlation_map(ndf)
```

### Top features that correlate with revenue are (Page value, exit rate and bounce rate, administrative, product related, special day and informational) however still need to encode month and vistor type and convert their values to numerical entities )

#### Encoding Months

```python
ndf["Month"].unique()
```

```python
d = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4,  'May':5, 'June':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
ndf.Month = ndf.Month.map(d)
```

```python
ndf['Month'].unique()
```

#### Encoding visitor type

```python
ndf["VisitorType"].unique()
```

```python
v={'Returning_Visitor':2, 'New_Visitor':1, 'Other':0 }
ndf.VisitorType = ndf.VisitorType.map(v)
```

#### Encoding weekend

```python
ndf["Weekend"].unique()
```

```python
ndf["Weekend"] = ndf["Weekend"].astype(int)
```

```python
plot_correlation_map(ndf)
```

## Dropping usless features and building models

```python
ndf.drop(['Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration', 'Browser', 'TrafficType'], axis=1,inplace = True)
```

```python
ndf
```

```python
ndf["Revenue"] = ndf["Revenue"].astype(int)
```

```python
ndf.info()
```

### Replacing NULL values

```python
ndf['Administrative'].fillna(df['Administrative'].mean(), inplace=True)
ndf['Informational'].fillna(ndf['Informational'].mean(), inplace=True)
ndf['ProductRelated'].fillna(ndf['ProductRelated'].mean(), inplace=True)
ndf['ProductRelated_Duration'].fillna(ndf['ProductRelated_Duration'].mean(), inplace=True)
ndf['Informational_Duration'].fillna(ndf['Informational_Duration'].mean(), inplace=True)
ndf['Administrative_Duration'].fillna(ndf['Administrative_Duration'].mean(), inplace=True)

ndf['BounceRates'].fillna(ndf['BounceRates'].mean(), inplace=True)
ndf['ExitRates'].fillna(ndf['ExitRates'].mean(), inplace=True)
```

```python
ndf.isnull().sum()
```

## Features selection 

```python
tdf=pd.read_csv('Test (1).csv')
x_test= ndf[['Administrative', 'Informational', 'Administrative_Duration', 'ProductRelated_Duration','ProductRelated',  'BounceRates',  'ExitRates', 'PageValues', 'SpecialDay', 'Month', 'VisitorType', 'Weekend' ]]
y_test= ndf['Revenue']
x= ndf[['Administrative', 'Informational', 'Administrative_Duration', 'ProductRelated_Duration','ProductRelated',  'BounceRates', 'ExitRates',  'PageValues', 'SpecialDay', 'Month', 'VisitorType', 'Weekend' ]]
y= ndf['Revenue']
```

### KNN

```python
#import the relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score

knn=KNeighborsClassifier(n_neighbors=18) #build our knn classifier
knn.fit(x, y) #Training KNN classifier
y_pred=knn.predict(x_test)  #Testing

print('Acuuracy=',accuracy_score(y_pred,y_test))
print('recall=',recall_score(y_pred,y_test))
print('f1 iscore=',f1_score(y_pred,y_test))
```

### XGBOOST

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(x, y)
y_pred = model.predict(x_test)

print('Acuuracy=',accuracy_score(y_pred,y_test))
print('recall=',recall_score(y_pred,y_test))
print('f1 score=',f1_score(y_pred,y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
